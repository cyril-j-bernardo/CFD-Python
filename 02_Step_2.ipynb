# Step 2: Non-linear Convection

Implementing non-linear convection on the same equations from last time:

$$\frac{\partial u}{\partial t} + u\frac{\partial u}{\partial x} = 0$$

With the difference being the *u* term, which used to be the constant, *c*. *u* in the second term makes the equation non-linear. We use the same discretization scheme as before:

$$\frac{u_i^{n+1} - u_i^n}{\Delta t} + u_i^n\frac{u_i^n - u_{i-1}^n}{\Delta x} = 0$$

Moving $u_i^{n+1}$ to one side of the equation:

$$u_i^{n+1} = u_i^n - u_i^n\frac{\Delta t}{\Delta x}(u_i^n - u_{i-1}^n)$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

num_x = 41
len_x = 2
dx = len_x / (num_x - 1)
num_t = 20
t = 0.5
dt = t / num_t

u = np.zeros([num_t, num_x]) # Initialize array of values
u[0, :] = 1
u[0, int(0.5/dx):int(1/dx + 1)] = 2

plt.plot(np.linspace(1, len_x, int(len_x/dx+1)), u[0,:])
```

```python
for i in range(1, num_t):
    for j in range(1, num_x):
        u[i, j] = u[i-1, j] - u[i-1, j] * (dt/dx) * (u[i - 1, j] - u[i, j - 1])
        
plt.plot(np.linspace(1, len_x, int(len_x/dx+1)), u[-1, :])
```
